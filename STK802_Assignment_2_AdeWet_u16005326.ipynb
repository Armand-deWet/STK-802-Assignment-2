# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
```

# Exploratory Data Analysis

```python
# import the csv to a pandas dataframe
claims_data = pd.read_csv('tdata.csv')
```

```python
# Quick descrption of the data
df_description = claims_data.describe()
df_description
```

```python
# Distribution of each feature and claim column
```

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = claims_data.columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(claims_data[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(claims_data[feature].skew(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.savefig("distributions.png", bbox_inches='tight') 
plt.show()
```

```python
# Box and Whisker plot of feature
```

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = claims_data.columns

# Plot Box and Whisker plot of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.boxplot(claims_data[feature], orient='h')
    plt.title(f"{feature}")

# Adjust layout and show plots
plt.tight_layout()
plt.savefig("box_whisker.png", bbox_inches='tight') 
plt.show()
```

```python
# Pair plot of features
```

```python
# Set the color palette
sns.set_palette("Pastel1")

# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 6))

# Using Seaborn to create a pair plot with the specified color palette
sns.pairplot(claims_data)

plt.suptitle('Pair Plot for Claims Data')
plt.savefig("pair_plot.png", bbox_inches='tight') 
plt.show()
```

```python
# Correlation Matrix of featues
```

```python
plt.figure(figsize=(15, 10))

# Using Seaborn to create a heatmap
sns.heatmap(claims_data.corr(), annot=True, fmt='.2f', linewidths=2)

plt.title('Correlation Heatmap of Claims Data')
plt.savefig("heat_map.png", bbox_inches='tight') 
plt.show()
```
